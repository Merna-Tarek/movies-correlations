```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Desktop/movies.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['gross']=df['gross'].astype('int64')
df['budget']=df['budget'].astype('int64')
```

```python
df
```

```python
df['yearcorrect']=df['released'].astype(str).str[:4]
```

```python
df
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
df.drop_duplicates()
```

```python
sns.scatterplot(data=df,x=df['budget'],y=df['gross'])
plt.xlabel("budget")
plt.ylabel('gross')
plt.title("Budget vs gross")
plt.show()
```

```python
sns.regplot(data=df,x=df['budget'],y=df['gross'])
```

```python
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
## normalize
df_nurmalized=df
for col_name in df_nurmalized.columns:
    if(df_nurmalized[col_name].dtype=='object'):
        df_nurmalized[col_name]=df_nurmalized[col_name].astype('category')
        df_nurmalized[col_name]=df_nurmalized[col_name].cat.codes
df_nurmalized.head()        
```

```python
correlation_matrix=df_nurmalized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
correlation_mat=df_nurmalized.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```
